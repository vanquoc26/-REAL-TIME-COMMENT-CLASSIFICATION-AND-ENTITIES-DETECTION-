```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
# Đọc dữ liệu từ file CSV (hoặc định dạng khác)
data_path = '/content/drive/MyDrive/Big Data/Data/output.csv' # Thay đổi đường dẫn tới file của bạn
df = pd.read_csv(data_path)

```

```python
df = df[df['xoa_khong'] != 1]
```

```python
df = df[['Comment', 'Label']]
df.head(5)
```

```python
df = df[df['Label'].isin([0, 1, 2])].reset_index(drop=True)
```

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm thử
X_train, X_test, y_train, y_test = train_test_split(df['Comment'], df['Label'], test_size=0.2, random_state=42)
```

```python
X_train = X_train.dropna()
X_test = X_test.dropna()
y_train = y_train[X_train.index]

vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score

# Dự đoán nhãn cho tập kiểm thử
y_pred = model.predict(X_test_vectorized)

# Đánh giá mô hình
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')  # Bạn có thể thay 'weighted' bằng 'macro', 'micro', hoặc 'binary' tùy theo mục đích của bạn

# In kết quả
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", report)
print(f"F1-Score: {f1}")
```
